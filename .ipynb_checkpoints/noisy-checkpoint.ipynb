```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fungsi untuk menambahkan noise salt and pepper ke gambar
def add_salt_and_pepper_noise(image, salt_probability, pepper_probability):
    output = np.copy(image)
    salt = np.ceil(salt_probability * image.size)
    pepper = np.ceil(pepper_probability * image.size)

    # Salt noise
    coords = [np.random.randint(0, i - 1, int(salt)) for i in image.shape]
    output[coords[0], coords[1]] = 255

    # Pepper noise
    coords = [np.random.randint(0, i - 1, int(pepper)) for i in image.shape]
    output[coords[0], coords[1]] = 0

    return output
```

```python
# Load image
akbar = cv2.imread('akbar.JPG', cv2.IMREAD_GRAYSCALE)
bagus = cv2.imread('bagus.JPG', cv2.IMREAD_GRAYSCALE)
hary = cv2.imread('hary.JPG', cv2.IMREAD_GRAYSCALE)
dihya = cv2.imread('dihya.JPG', cv2.IMREAD_GRAYSCALE)
miftah = cv2.imread('miftah.JPG', cv2.IMREAD_GRAYSCALE)
```

```python
# Menambahkan noise salt and pepper ke setiap gambar
salt_probability = 0.2  # Sesuaikan sesuai preferensi Anda
pepper_probability = 0.2  # Sesuaikan sesuai preferensi Anda
```

```python
akbar_noisy = add_salt_and_pepper_noise(akbar, salt_probability, pepper_probability)
bagus_noisy = add_salt_and_pepper_noise(bagus, salt_probability, pepper_probability)
hary_noisy = add_salt_and_pepper_noise(hary, salt_probability, pepper_probability)
dihya_noisy = add_salt_and_pepper_noise(dihya, salt_probability, pepper_probability)
miftah_noisy = add_salt_and_pepper_noise(miftah, salt_probability, pepper_probability)
```

```python
# Menampilkan gambar-gambar yang telah ditambahkan noise
plt.figure(figsize=(10, 6))

plt.subplot(231)
plt.imshow(akbar_noisy, cmap='gray')
plt.title('Akbar Noisy')
plt.axis('off')

plt.subplot(232)
plt.imshow(bagus_noisy, cmap='gray')
plt.title('Bagus Noisy')
plt.axis('off')

plt.subplot(233)
plt.imshow(hary_noisy, cmap='gray')
plt.title('Hary Noisy')
plt.axis('off')

plt.subplot(234)
plt.imshow(dihya_noisy, cmap='gray')
plt.title('Dihya Noisy')
plt.axis('off')

plt.subplot(235)
plt.imshow(miftah_noisy, cmap='gray')
plt.title('Miftah Noisy')
plt.axis('off')

plt.show()
```

```python
cv2.imwrite('akbar_noisy.jpg', akbar_noisy)
cv2.imwrite('bagus_noisy.jpg', bagus_noisy)
cv2.imwrite('hary_noisy.jpg', hary_noisy)
cv2.imwrite('dihya_noisy.jpg', dihya_noisy)
cv2.imwrite('miftah_noisy.jpg', miftah_noisy)
```
