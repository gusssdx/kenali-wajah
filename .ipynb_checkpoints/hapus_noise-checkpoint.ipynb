```python
import cv2
from scipy.signal import convolve2d
from scipy.ndimage import gaussian_filter
import numpy as np
from matplotlib import pyplot as plt
```

```python
def wiener_filter(image, kernel_size=(6, 6), noise_variance=0.01):
    kernel = np.ones(kernel_size) / np.prod(kernel_size)
    image_mean = convolve2d(image, kernel, mode='same')
    image_var = convolve2d(image**2, kernel, mode='same') - image_mean**2
    noise_var = noise_variance
    k = noise_var / (noise_var + image_var)
    restored_image = image_mean + k * (image - image_mean)
    return np.clip(restored_image, 0, 255).astype(np.uint8)
```

```python
def histogram_equalization(img):
    histogram, bins = np.histogram(img.flatten(), bins=256, range=(0,256))
    cdf = histogram.cumsum()
    cdf_normalized = cdf / cdf.max()
    equalized_histogram = (cdf_normalized * 255).astype('uint8')
    equalized_image = equalized_histogram[img]
    return equalized_image
```

```python
def median_filter(img, kernel_size):
    
    height, width = img.shape

    # Ukuran kernel
    kernel_height, kernel_width = kernel_size

    # Padding untuk mengatasi border
    pad_height = kernel_height // 2
    pad_width = kernel_width // 2
    filtered_image = np.zeros((height, width), dtype=np.uint8)

    # Iterasi melalui setiap piksel dalam gambar
    for i in range(pad_height, height - pad_height):
        for j in range(pad_width, width - pad_width):
            # Ambil nilai piksel dalam kernel
            neighborhood = img[i - pad_height:i + pad_height + 1, j - pad_width:j + pad_width + 1]

            # Ambil nilai median dari neighborhood
            median_value = np.median(neighborhood)

            # Simpan nilai median ke dalam gambar hasil
            filtered_image[i, j] = median_value

    return filtered_image
```

```python
# Load image
akbar = cv2.imread('akbar_noisy.jpg', cv2.IMREAD_GRAYSCALE)
bagus = cv2.imread('bagus_noisy.jpg', cv2.IMREAD_GRAYSCALE)
hary = cv2.imread('hary_noisy.jpg', cv2.IMREAD_GRAYSCALE)
dihya = cv2.imread('dihya_noisy.jpg', cv2.IMREAD_GRAYSCALE)
miftah = cv2.imread('miftah_noisy.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Ukuran kernel
kernel_size = (5, 5)
```

```python
# Apply Wiener filter and histogram equalization
wn_akbar = wiener_filter(akbar)
wn_hs_akbar = histogram_equalization(wn_akbar)
# Apply median filter
md_akbar = median_filter(akbar, kernel_size)
```

```python
display = [akbar, wn_akbar,  wn_hs_akbar, md_akbar]
label = ['noisy Image', 'wiener filter','wiener + histogram filter', 'median filter']

fig = plt.figure(figsize=(12, 10))

for i in range(len(display)):
    fig.add_subplot(2, 2, i+1)
    plt.imshow(display[i], cmap = 'gray')
    plt.title(label[i])

plt.show()
```
